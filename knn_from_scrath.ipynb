```python
import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
```

```python
X = np.array([[1,80],
             [2,78],
             [3,55],
             [7,71],
             [10,62]])
y = np.array(["Pass","Pass","Fail","Pass","Fail"])

```

```python
#DEFINE THE DISTANCE BETWEEN TWO POINTS (Arrays)
def dist (a,b):
    return np.sqrt((a[0]-b[0])**2+(a[1] - b[1])**2)
```

```python
def knn (X,y,new_pt,k=3):
    distance = []
    for i in range(len(X)):
        d = dist(new_pt,X[i])
        distance.append((d,y[i]))
    distance.sort(key=lambda x:x[0])
    k_nearest = distance[:k]
    labels = [label for _,label in k_nearest]
    predict = Counter(labels).most_common(1)[0][0]
    return predict
predct = knn(X,y,np.array([15,71]),3)
print (predct)

        
```
