```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create some data values for both X[predictor] and Y[target]
X = np.array([[1,29],[2,30],[3,11],[4,15],[5,6]])
y = np.array(["top","top",'fail','top','fail'])
```

```python
# KNN FUNCTION

def knn_function (X,y,new_pt,k):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X,y)
    prediction= knn.predict(new_pt)[0]
    
    x_point = []
    y_point = []
    for i in range (len(X)):
        x_value = X[i][0]
        y_value = X[i][1]
        x_point.append(x_value)
        y_point.append(y_value)
    plt.scatter(x_point,y_point,)
    plt.scatter(new_pt[0][0],new_pt[0][1])
    return prediction
   


```

```python
out = knn_function(X,y,np.array([[10,30]]),3)
print(f'predicted value is {out}')
```
