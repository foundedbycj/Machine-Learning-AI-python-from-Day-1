```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
X = np.array([[1,25],[2,30],[3,29],[4,12],[5,5],[6,18]])
y = np.array(["pass","pass","pass","fail","fail","pass"])
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X,y)
predict = knn.predict(np.array([[12,23]]))
predict=predict[0]
print (predict)

```
